```python
import numpy as np
x,y=np.array([2.5,.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1]),np.array([2.4,.7,2.9,2.2,3.,2.7,1.6,1.1,1.6,.9])
n=len(x)
```

```python
x_mean,y_mean=np.mean(x),np.mean(y)
new_x,new_y=x-x_mean,y-y_mean
new_data=(new_x,new_y)
```

```python
cov_x_x=np.sum((x-x_mean)**2)/(n-1)
cov_x_y=np.sum((x-x_mean)*(y-y_mean))/(n-1) #cove_x_y=cov_y_x(same)
cov_y_y=np.sum((y-y_mean)**2)/(n-1)
cov_x_x,cov_x_y,cov_y_y
```

```python
cov_metrix=np.array([[cov_x_x,cov_x_y],[cov_x_y,cov_y_y]])
cov_metrix
```

```python
eigenvalues,eigenvectors=np.linalg.eig(cov_metrix)
eigenvalues,eigenvectors
```

```python
if eigenvalues[0]>eigenvalues[1]:
  feature_vector=eigenvectors[0]
else:
  feature_vector=eigenvectors[1]
feature_vector
```

```python
Final_data=feature_vector@new_data
Final_data
```
