```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
```

```python
x,y=[6,8,12,14,18],[350,775,1150,1395,1675]
```

```python
x_avg,y_avg,xy_avg,x_sqr_avg=0,0,0,0
size=len(x)
for i in range (size):
  x_avg+=x[i]
  y_avg+=y[i]
  xy_avg+=x[i]*y[i]
  x_sqr_avg+=x[i]*x[i]

x_avg,y_avg,xy_avg,x_sqr_avg=x_avg/size,y_avg/size,xy_avg/size,x_sqr_avg/size
x_avg_sqr=x_avg*x_avg
x_avg,y_avg,xy_avg,x_sqr_avg,x_avg_sqr
```

```python
m=(x_avg*y_avg-xy_avg)/(x_avg_sqr-x_sqr_avg)
c=y_avg-m*x_avg
m,c
```

```python
y_pred=[]
for i in range (size):
  y_pred.append(m*x[i]+c)
y_pred
```

```python
up_sqr,dn_sqr=0,0
for i in range (size):
  up_sqr+=(y_pred[i]-y_avg)*(y_pred[i]-y_avg)
  dn_sqr+=(y[i]-y_avg)*(y[i]-y_avg)
r_sqr=up_sqr/dn_sqr
r_sqr
```

```python
print(str(r_sqr * 100) + '%')
```
