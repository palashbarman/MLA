```python
import numpy as np
x=np.array([[2,4],[2,3],[5,2],[6,2],[5,2.5],[2.5,3.5]])
n=len(x)
m1=x[0].copy()
m2=x[n-1].copy()
cluster1,cluster2,a,b,c,d=[],[],[],[],[],[]
```

```python
for i in range(n):
  dis1=np.sqrt(np.sum((m1-x[i])**2))
  dis2=np.sqrt(np.sum((m2-x[i])**2))
  if dis1<dis2:
    cluster1.append(x[i].copy())
    a.append(x[i][0])
    b.append(x[i][1])
    m1[0],m1[1]=np.mean(a),np.mean(b)
  else:
    cluster2.append(x[i].copy())
    c.append(x[i][0])
    d.append(x[i][1])
    m2[0],m2[1]=np.mean(c),np.mean(d)
print(np.array(cluster1))
print(np.array(cluster2))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Sample data
X, _ = make_blobs(n_samples=300, centers=5, random_state=42)

# Run KMeans with k from 1 to 10
wss = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wss.append(kmeans.inertia_)  # Inertia is WSS

# Plotting the Elbow Method
plt.plot(k_range, wss, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WSS)')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```
