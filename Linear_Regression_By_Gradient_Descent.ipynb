```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x,y=[6,8,12,14,18],[350,775,1150,1395,1675]
```

```python
m,c,l,iteration,n=0,0,.001,50000,len(x)
for i in range(iteration):
  y_pred,der_m,der_c=[],0,0
  for j in range(n):
    y_pred.append(m*x[j]+c)
    der_m+=(y[j]-y_pred[j])*x[j]
    der_c+=(y[j]-y_pred[j])
  m=m+l*der_m/n
  c=c+l*der_c/n
m,c
```

```python
Y=np.mean(y)
up=np.subtract(y_pred,Y)
dn=np.subtract(y,Y)
up*=up
dn*=dn
ups,dns=np.mean(up),np.mean(dn)
r_sqrt=ups/dns
r_sqrt
```

```python
print(r_sqrt*100,'%')
```
