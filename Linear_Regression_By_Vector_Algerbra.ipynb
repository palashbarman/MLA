```python
import pandas as pd
import numpy as np
```

```python
x=np.array([1,2,4,3,5])
y=np.array([1,3,3,2,5])
x=np.column_stack((np.ones(len(x)),x))
x.shape
```

```python
x=np.array([6,8,12,14,18])
y=np.array([350,775,1150,1395,1675])
x=np.column_stack((np.ones(len(x)),x))
x.shape
```

```python
w=np.linalg.inv(x.T@x)@x.T@y
w
```

```python
predict=x@w
predict
```

```python
Y=np.mean(y)
r_sqr=np.sum((predict-Y)**2)/np.sum((y-Y)**2)
print(r_sqr*100,'%')
```
