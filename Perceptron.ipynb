```python
import numpy as np
x,y=np.array([[0,0],[0,1],[1,0],[1,1]]),np.array([0,0,0,1])
m,n=x.shape
```

```python
#w=np.zeros(n)
w=np.array([0.3,-.1])
alpha,th,epoch,i,a=.1,.2,500,0,0
while(i<=epoch):
  if a==m:
    break
  a=0
  for j in range(m):
    dot=round(w@x[j],10)
    if (dot)>=th:
      yp=1
    else:
      yp=0
    error=y[j]-yp
    if error==0:
      a+=1
    w=w+alpha*x[j]*error
    #print(dot,error,w)
  i+=1
w,a,i
```
