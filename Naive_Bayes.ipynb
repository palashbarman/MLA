```python
import numpy as np
x=np.array([['Sunny','Cold','Indoor'],['Sunny','Warm','Outdoor'],['Cloudy','Warm','Indoor'],['Sunny','Warm','Indoor'],['Cloudy','Cold','Indoor'],['Cloudy','Cold','Outdoor'],['Sunny','Cold','Outdoor']])
y=np.array(['no','no','no','no','yes','yes','yes'])
m,n=x.shape
```

```python
out_s,out_c,tem_w,tem_co,rout_o,rout_i=[],[],[],[],[],[]
output_yes=0
for i in range(m):
  if x[i][0]=='Sunny':
    out_s.append([x[i][0],y[i]])
  else:
    out_c.append([x[i][0],y[i]])
  if x[i][1]=='Warm':
    tem_w.append([x[i][1],y[i]])
  else:
    tem_co.append([x[i][1],y[i]])
  if x[i][2]=='Outdoor':
    rout_o.append([x[i][2],y[i]])
  else:
    rout_i.append([x[i][2],y[i]])
  if y[i]=='yes':
    output_yes+=1
output_no=len(y)-output_yes


out_s,out_c,tem_w,tem_co,rout_o,rout_i=np.array(out_s),np.array(out_c),np.array(tem_w),np.array(tem_co),np.array(rout_o),np.array(rout_i)
print(out_s,out_c,tem_w,tem_co,rout_o,rout_i)
```

```python
def outlook(t):
  if t=="Sunny":
    out_s_yes=0
    for i in range(len(out_s)):
      if(out_s[i][1]=='yes'):
        out_s_yes+=1
    out_s_no=len(out_s)-out_s_yes
    pout_s_yes=out_s_yes/output_yes
    pout_s_no=out_s_no/output_no
    return pout_s_yes,pout_s_no,len(out_s)/m
  else:
    out_c_yes=0
  for i in range(len(out_c)):
    if(out_c[i][1]=='yes'):
      out_c_yes+=1
  out_c_no=len(out_c)-out_c_yes
  pout_c_yes=out_c_yes/output_yes
  pout_c_no=out_c_no/output_no
  return pout_c_yes,pout_c_no,len(out_c)/m
```

```python
def temperature(t):
  if t=="Warm":
    tem_w_yes=0
    for i in range(len(tem_w)):
      if(tem_w[i][1]=='yes'):
        tem_w_yes+=1
    tem_w_no=len(tem_w)-tem_w_yes
    ptem_w_yes=tem_w_yes/output_yes
    ptem_w_no=tem_w_no/output_no
    return ptem_w_yes,ptem_w_no,len(tem_w)/m
  else:
    tem_co_yes=0
    for i in range(len(tem_co)):
      if(tem_co[i][1]=='yes'):
        tem_co_yes+=1
    tem_co_no=len(tem_co)-tem_co_yes
    ptem_co_yes=tem_co_yes/output_yes
    ptem_co_no=tem_co_no/output_no
    return ptem_co_yes,ptem_co_no,len(tem_co)/m
```

```python
def routine(t):
  if t=="Outdoor":
    rout_o_yes=0
    for i in range(len(rout_o)):
      if(rout_o[i][1]=='yes'):
        rout_o_yes+=1
    rout_o_no=len(rout_o)-rout_o_yes
    prout_o_yes=rout_o_yes/output_yes
    prout_o_no=rout_o_no/output_no
    return prout_o_yes,prout_o_no,len(rout_o)/m
  else:
    rout_i_yes=0
    for i in range(len(rout_i)):
      if(rout_i[i][1]=='yes'):
        rout_i_yes+=1
    rout_i_no=len(rout_i)-rout_i_yes
    prout_i_yes=rout_i_yes/output_yes
    prout_i_no=rout_i_no/output_no
    return prout_i_yes,prout_i_no,len(rout_i)/m
```

```python
poutput_yes=output_yes/m
poutput_no=output_no/m
```

```python
test=np.array(['Cloudy','Warm','Outdoor'])
#P(x|c)=P(c|x)*P(x)/P(c)
#P(c=yes|x)=P(x|c=yes)*P(c=yes)/P(x)
#P(c=no|x)=P(x|c=no)*P(c=no)/P(x)
t0,t1,t2=outlook(test[0]),temperature(test[1]),routine(test[2])
P_test_Yes=(t0[0]*t1[0]*t2[0]*poutput_yes)/(t0[2]*t1[2]*t2[2])
P_test_no=(t0[1]*t1[1]*t2[1]*poutput_no)/(t0[2]*t1[2]*t2[2])
print(P_test_Yes,P_test_no,t0,t1,t2)
```

```python
if P_test_Yes> P_test_no:
  print("Result will be yes for :",test)
else:
  print("Result will be no for :",test)
```
